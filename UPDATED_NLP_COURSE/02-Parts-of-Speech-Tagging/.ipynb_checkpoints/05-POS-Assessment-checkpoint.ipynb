# Parts of Speech Assessment

For this assessment we'll be using the short story [The Tale of Peter Rabbit](https://en.wikipedia.org/wiki/The_Tale_of_Peter_Rabbit) by Beatrix Potter (1902). <br>The story is in the public domain; the text file was obtained from [Project Gutenberg](https://www.gutenberg.org/ebooks/14838.txt.utf-8).

```python
# RUN THIS CELL to perform standard imports:
import spacy
nlp = spacy.load('en_core_web_sm')
from spacy import displacy
```

**1. Create a Doc object from the file `peterrabbit.txt`**<br>
> HINT: Use `with open('../TextFiles/peterrabbit.txt') as f:`

```python
with open('../TextFiles/peterrabbit.txt') as f:
    doc = nlp(f.read())
```

```python
count = 0
sentence = ""
for sent in doc.sents:
    count += 1
    if count == 3:
        sentence = sent
        break
for token in sentence:
    print(f'{token.text} {token.pos_} {token.tag_} {spacy.explain(token.tag_)}')
```

**2. For every token in the third sentence, print the token text, the POS tag, the fine-grained TAG tag, and the description of the fine-grained tag.**

```python
# Enter your code here:


```

```python
from collections import Counter
pos_counter = Counter()
for token in doc:
    pos_counter[token.pos_] += 1
# Get the counts of POS tags
POS_counts = doc.count_by(spacy.attrs.POS)
# Iterate over the POS tags, sort them, and display the frequency list
for pos_id, count in sorted(POS_counts.items()):
    pos_name = doc.vocab[pos_id].text
    print(f'{pos_id}. {pos_name:{5}}: {count}')
```

**3. Provide a frequency list of POS tags from the entire document**

**4. CHALLENGE: What percentage of tokens are nouns?**<br>
HINT: the attribute ID for 'NOUN' is 91

```python
print(len(doc))
for pos_id, count in sorted(POS_counts.items()):
    pos_name = doc.vocab[pos_id].text
    if pos_name == "NOUN":
        percentage = (count / len(doc)) * 100
        print(f'{count}/{len(doc)} = {percentage:.2f}%')
```

**5. Display the Dependency Parse for the third sentence**

```python
from spacy import displacy
displacy.render(sentence, style='dep', jupyter=True)
```

**6. Show the first two named entities from Beatrix Potter's *The Tale of Peter Rabbit* **

```python
named_entities = []
count = 0

# Iterate through named entities in the document
for ent in doc.ents:
    label = ent.label_
    description = spacy.explain(label)
    named_entities.append(f"{ent.text} - {label} - {description}")
    count += 1
    
    # Break after the first two named entities
    if count == 2:
        break

# Print the first two named entities
for entity in named_entities:
    print(entity)
```

**8. CHALLENGE: How many sentences contain named entities?**

```python
docu2 = nlp("Beatrix Potter was a famous author. She lived in London.")

# Extract Span objects representing named entities
named_entities2 = [ent for ent in docu2.ents]
print(named_entities2)

for entity in named_entities2:
    if type(entity) == spacy.tokens.span.Span:
        print("It's a Span object")
    else:
        print("It's not a Span object")
```

```python
sentences_with_entities = 0
first_approach = []

# Iterate through sentences in the document
for sent in doc.sents:
    # Check if the sentence contains at least one named entity
    if any(ent.label_ for ent in sent.ents):
        first_approach.append(nlp(sent.text))
        sentences_with_entities += 1

for entity in first_approach:
    if type(entity) == spacy.tokens.span.Span:
        print("It's a Span object")
    else:
        print("It's not a Span object")        

# Print the count of sentences with named entities
print(f"Number of sentences with named entities: {sentences_with_entities}")

list_of_sents = [nlp(sent.text) for sent in doc.sents]
list_of_ners = [doc for doc in list_of_sents if doc.ents]
print(len(list_of_ners))

for entity in list_of_ners:
    if type(entity) == spacy.tokens.doc.Doc:
        print("It's a Doc object")
    else:
        print("It's not a Doc object")

for sent1, sent2 in zip(first_approach, list_of_ners):
    print(f"From first_approach: {sent1.text}")
    print(f"From list_of_ners: {sent2.text}")
    print()
```
